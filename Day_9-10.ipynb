```python
# Dealing with missing data
```

```python
import pandas as pd
import numpy as np
```

```python
from numpy import NaN
```

```python
data1 = {'Name':['person1','person2','person3','person4'],'Age':[28,27,NaN,36]}
print(data1)
```

```python
data = pd.DataFrame(data1)
data
```

```python
data.isnull() #if any value is null it returns True
```

```python
data.isnull().sum() #return total null values in each column
```

```python
data.dropna() #rows containing null values will be removed and new copy will be displayed
# data.dropna(inplace=True) #permanent change
```

```python
data['Age'].dtype #float because NaN is given max memory if new values in future are given float values
```

```python
# filling missing values with mean/median/mode
data['Age'].fillna(data['Age'].mean())
#data['Age'].fillna(data['Age'].mean(),inplace=True)
```

```python
# using median
data['Age'].fillna(data['Age'].median())
```

```python
# creating multiple subplots in a graph
x1_axis = [2,4,6,8]
y1_axis = [1,10,100,1000]

x2_axis = [1,3,5,7]
y2_axis = [100,110,130,120]

plt.subplot(1,2,1) #plt.subplot(no. of rows, no. of columns, current plot no)
plt.plot(x1_axis,y1_axis,'r')

plt.subplot(1,2,2)
plt.plot(x2_axis,y2_axis,'g')    

plt.tight_layout()
    
```

```python
plt.subplot(2,1,1) #plt.subplot(no. of rows, no. of columns, current plot no)
plt.plot(x1_axis,y1_axis,'r')

plt.subplot(2,1,2)
plt.plot(x2_axis,y2_axis,'g')    

plt.tight_layout()
```

```python
plt.subplot(2,1,1)  #title or label should be in b/w subplot & plot
plt.title("Exponential")
plt.plot(x1_axis,y1_axis,'r')

plt.subplot(2,1,2)
plt.title("Linear")
plt.plot(x2_axis,y2_axis,'g')    

plt.tight_layout()
```
